# Installation

1. Cloner le dépôt: git clone [URL]
2. Exécuter setup.bat pour installer l'environnement Python
3. Exécuter download_llama.bat pour télécharger llama.cpp
4. Activer l'environnement: venv\Scripts\activate
5. Exécuter python download_model.py pour télécharger le modèle

# Démarrage des services

1. Démarrer le serveur llama.cpp: start_server.bat (dans un terminal)
2. Démarrer n8n: start_n8n.bat (dans un autre terminal)
3. Démarrer Jupyter: start_jupyter.bat (dans un troisième terminal)
